```python
class Polynomial:
    def __init__(self, *coeff):
        self.coeff = coeff
    def __repr__(self):
        return 'Polynomial *{0!r}'.format(self.coeff, 2)
    
    def __add__(self, other):
        return Polynomial(*(x+y for x,y in zip(self.coeff, other.coeff)))

    def __len__(self):
        return len(self.coeff) -1
    
p1 = Polynomial(1,2,3) #x^2 + 2x +3
p2 = Polynomial(3,4,3) #3x^2 + 4x + 3
```

```python
#2 meta class :- how the methods are constructed 
# even though we only have the option to touch the base classes we can make difference and can check what the derieved classes contains

class MetaClass(type):
    def __new__(cls, name, base, body):
        print(f"MetaClass:::-\ncls:- {cls} \nname:- {name} \nbase:- {base} \nbody:- {body}")
        if "bar" in body:
            print("method bar exist in the derived class")
        else:
            print("method bar does not exit")
        return super().__new__(cls,name, base, body)
    
class Base(metaclass=MetaClass):
    def foo(self):
        return self.bar()
##################################################################################     
class Derived(Base):
    def bar(self):
        print ("bar")
    def bar1(self):
        print("bar1")
```

```python
# 3. __init_subclass
# tried to get the function name using __init_subclass__() but it is not present in it,
# may be some other function have it

class MetaClass(type):
    def __new__(cls, name, base, body, **kw):
        print(f"MetaClass:::-\ncls:- {cls} \nname:- {name} \nbase:- {base} \nbody:- {body}")
        if "bar" in body:
            print("method bar exist in the derived class\n\n")
        else:
            print("method bar does not exit\n\n")
        return super().__new__(cls,name, base, body)
    
class Base(metaclass=MetaClass):
    def foo(self):
        return self.bar()
    
    def __init_subclass__(cls, **kw):
        print("init_subclass:- \n", cls)
        return super().__init_subclass__( **kw)
    
##################################################################################     
class Derived(Base, fun1='bar', fun2="bar1"):
    def bar(self):
        print ("bar")
    def bar1(self):
        print("bar1")
```

```python
from time import time

def add(x, y=10):
    return x + y

def sub(x, y=9):
    return x-y

# what if we want to calcualte the time elapse in each execution of these functions

before = time()
print("a(10)", add(10))
after = time()
print("time elapse ", after - before)

before = time()
print("a(10, 20)", add(10, 20))
after = time()
print("time elapse ", after - before)

before = time()
print("a('a', 'b')", add('a', 'b'))
after = time()
print("time elapse ", after - before)

before = time()
print("sub(10)", sub(10))
after = time()
print("time elapse ", after - before)
```

```python
# previous one way is the rough way of doing the things or just say do the thing any how
# this way is slightly better way 

from time import time

def timer(fun, x, y=0):
    before = time()
    rv = fun(x,y)
    after = time()
    print("time elapse = ", after - before)
    return rv
    
def add(x, y=10):
    return x + y

def sub(x, y=9):
    return x-y

print("a(10)", timer(add,10))
print("a(10, 20)", timer(add,10, 20))
print("a('a', 'b')", timer(add,'a', 'b'))
print("sub(10)", timer(sub,10))
```

```python
# but we can do better than previous one as well
# currently we are changing the name of the function what if end user do't know this thing.
# just arranging the previous same code and implementing it as a wrapper 

from time import time

def timer(fun):
    def f(x, y=0):
        before = time()
        rv = fun(x,y)
        after = time()
        print("time elapse = ", after - before)
        return rv
    return f
    
@timer
def add(x, y=10):
    return x + y

# add = timer(add)  ==> @timer

@timer
def sub(x, y=9):
    return x-y

# sub = timer(sub)  ==> @timer 

print("a(10)", add(10))
print("a(10, 20)", add(10, 20))
print("a('a', 'b')", add('a', 'b'))
print("sub(10)", sub(10))
```

```python
#decorators

from time import time

def timer(fun):
    def f(x, y=0):
        before = time()
        rv = fun(x,y)
        after = time()
        print("time elapse = ", after - before)
        return rv
    return f
    
@timer
def add(x, y=10):
    return x + y

@timer
def sub(x, y=9):
    return x-y

print("a(10)", add(10))
print("a(10, 20)", add(10, 20))
print("a('a', 'b')", add('a', 'b'))
print("sub(10)", sub(10))
```
